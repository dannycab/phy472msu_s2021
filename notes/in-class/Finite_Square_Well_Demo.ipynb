## Finite Square Well Demo

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f1(z):
    
    return z*np.tan(z)

def f2(z):
    
    return -z/np.tan(z)

def g(z,z0):
    
    return np.sqrt(z0**2-z**2)

def h1(z,z0):
    
    return g(z,z0)-f1(z)

def h2(z,z0):
    
    return g(z,z0)-f2(z)
```

```python
z0 = 10
zarr = np.arange(0,z0,0.01)

plt.plot(zarr, f1(zarr))
plt.plot(zarr, f2(zarr))
plt.plot(zarr, g(zarr,z0))

plt.axis([0,z0,0,z0])
```
